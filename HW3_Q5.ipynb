```python
import numpy as np
import cvxpy as cp
```

```python
def grad_F(x):
    return 2*x/(1 - np.dot(x,x))
def hessian_F(x):
    a = np.zeros((20,20))

    for i in range(20):
        for j in range(20):
            if i == j:
                a[i,j] = 2*(1 + np.dot(x,x))/(1 - np.dot(x,x))**2
            else:
                a[i,j] = (4*x[i]*x[j])/(1 - np.dot(x,x))**2
    return a

    # print(4/((1 - np.dot(x,x))**2))
    # return 2/(1 - np.dot(x,x)) + 4/((1 - np.dot(x,x))**2)*a
def local_norm(g,x):
    # print(hessian_F(x))
    hessian_inv = np.linalg.inv(hessian_F(x))
    return np.sqrt(g.T@hessian_inv@g)
```

```python
c = np.ones(20)
ITERATION = 1000
t = np.zeros(ITERATION)
y = np.zeros((20,ITERATION))
# y[:,0] = -0.3*np.ones(20)
epsilon = np.zeros(ITERATION)
lda = np.zeros(ITERATION)
beta = 1/np.sqrt(2)
gamma = 1/np.sqrt(2)
v = 2
```

```python
for k in range(ITERATION):
    # print(f'Working with k = {k}')
    # print('Updating t')
    t[k + 1] = t[k] + gamma/local_norm(c,y[:,k])
    update_val = t[k+1]*c + grad_F(y[:,k])
    # print('Calculateing lambda')
    lda[k] = local_norm(update_val, y[:,k])
    # print('Calculating epsilon')
    epsilon[k] = (lda[k]**2)/(1 + lda[k])
    # print('Updating y')
    update = 1/(1 + epsilon[k])*np.linalg.inv(hessian_F(y[:,k]))@(update_val)
    y[:,k + 1] = y[:,k] - update
    if t[k] >= 1/0.1*(v + (beta +np.sqrt(v))*beta/(1 - beta)):
        print(k)
        break
```

```python
x = cp.Variable(20)
obj = cp.Minimize(np.ones(20)@x)
constraints = [x <= 1, x >= -1]
prob = cp.Problem(obj, constraints)
prob.solve()
```
